```python
import pandas as pd
#Method 1: Manual Data Entry
#collected survey data on campus about Study Hours vs Marks
data = {
    'Student_Name': ['Rajiv', 'Mohan', 'Rohan', 'Aditya', 'Vikrant'],
    'Study_Hours': [6, 9, 10, 12, 8],
    'Marks': [88, 90, 100, 99, 97]
}
df_manual = pd.DataFrame(data)
# saving csv file
df_manual.to_csv('manual_data.csv', index=False)
print("Success! 'manual_data.csv' has been created.")
df_manual
```

```python
import requests
from bs4 import BeautifulSoup
# Method 2: Web Scraping
#Extracting book titles from a retailer's website
url = 'http://books.toscrape.com/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
# Find all books on the page
books = soup.find_all('article', class_='product_pod')
titles = []
for book in books:
    # Get the title from the HTML
    title = book.h3.a['title']
    titles.append(title)
# Save the list to a CSV file
df_scrape = pd.DataFrame({'Book_Title': titles})
df_scrape.to_csv('scraped_data.csv', index=False)
print("Success! 'scraped_data.csv' has been created.")
df_scrape
```

```python
# Method 3: API Collection
#Getting live weather data from a server
url = "https://api.open-meteo.com/v1/forecast?latitude=20.296&longitude=85.824&current_weather=true"
response = requests.get(url)
data = response.json()
# Extract just the current weather part
weather = data['current_weather']
# Save it to a CSV
df_api = pd.DataFrame([weather])
df_api.to_csv('api_data.csv', index=False)
print("Success! 'api_data.csv' has been created.")
df_api
```
